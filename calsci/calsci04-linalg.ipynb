# Linalg (linear algebra)

Numpy intègre le calcul matriciel (ou l'algèbre linéaire) dans sa sous-bibliothèque [numpy.linalg](https://docs.scipy.org/doc/numpy/reference/routines.linalg.html). Pour être efficace il est
important que Numpy soit relié aux bibliothèques [Lapack](https://www.netlib.org/lapack/) et [BLAS](https://www.netlib.org/blas/) (la version d'Intel est [MKL](https://www.intel.com/content/www/us/en/developer/tools/oneapi/onemkl.html)). Ces bibliothèques sont imbatables, Numpy relié à ces bibliothèques sera largement plus rapide qu'un programme dans n'importe quel autre langage qui ne les utilise pas (vous relevez le défi ?).

La bibliothèque [Scipy](https://scipy.org/) a aussi une sous-bibliothèque [linalg](https://docs.scipy.org/doc/scipy/reference/linalg.html) qui est très proche. Si vous ne trouvez pas
ce que vous cherchez dans la version de Numpy, il peut être intéressant de regarder si Scipy l'a.

```python
import numpy as np
import numpy.linalg as lin

np.set_printoptions(precision=3)
np.show_config()
```

## Opérations de base

On a vu que les opérateurs +, -, \* et /  sont appliqués terme à terme ce qui est juste pour + et - dans le cadre du calcul matriciel mais pas pour \* et /.

* Le __produit scalaire__ utilise la méthode `dot` ou l'opérateur `@`
* La division que l'on peut imaginer comme 
   * la __résolution__ d'un système matriciel utilise la fonction `solve`
   * le calcul de l'__inverse__ de la matrice utilise la fonction `inv`

```python
A = np.array([[1,2],[3,4]])
print("multiplication terme à terme : \n",A * A) # tous les opérateurs sont appliqués terme à terme
print("produit matriciel : \n", A.dot(A))        # A @ A can be used too
```

Résolution de système matriciel :

```python
b = np.array([17,37])
x = lin.solve(A, b) # bien mieux que de calculer la matrice inverse (plus rapide et plus stable)
print("x = ", x)
print("verification : ", A @ x)
```

Si on le désire vraiment, on peut calculer la matrice inverse (mais c'est plus long) :

```python
print("A⁻¹ :\n", lin.inv(A))       # la matrice inverse
print("x = ", lin.inv(A).dot(b))
```

Enfin __la transposée__ est simplement `T` :

```python
A.T
```

## Extractions

On peut récupérer

* la diagonale d'une matrice avec la fonction `diag` (attention le résultat est un vecteur si l'argument est une matrice et une matrice si l'argument est un vecteur)
* la matrice triangulaire inférieur avec la fonction `tril` (l pour lower) et supérieure avec `triu` (u pour upper). Il est possible de décaler la diagonale du triangle, voir la doc

On peut aussi construire une matrice triangulaire avec des 1 et 0 avec `tri`et donc aussi une matrice triangulaire quelconque :

```python
np.tri(3,3) * np.random.randint(1, 10, size=(3,3))
```

Et voici comment extraire une matrice tridiagonale :

```python
A = np.random.randint(1, 10, size=(5,5))
np.tril(np.triu(A, k=-1), k=1)
```

## Opérations sur matrices

La bibliothèque offre des fonctions de

* décomposition (LU, Choleski, QR, SVD...)
* valeurs et vecteurs propres
* norme, déterminant, conditionnement et rang

```python
Q,R = lin.qr(A)   # Q est orthogonale, R est triangulaire supérieur
print(Q, '\n\n', R)
print("\nVérification :\n", Q.dot(R))
```

#### Valeurs propres et vecteurs propres

```python
lin.eig(A) # donne les valeurs propres et les vecteurs propres de A
```

#### Déterminant, norme etc. 

```python
print("Déterminant :", lin.det(A))
print("Norme 2 :", lin.norm(A), "\nNorme 1 :", lin.norm(A, 1) )
print("Conditionnement :", lin.cond(A,2))  # I choose norm 2 to compute the condition number
```
