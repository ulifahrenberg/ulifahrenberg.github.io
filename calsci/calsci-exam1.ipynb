### Postscriptum

La suite de Perrin satisfait la propriété que *si $n$ est un nombre premier, alors $x_n$ est divisible par $n$*.
La réciproque de cela n'est **pas vrai**, mais le plus petit contre-exemple se trouve à $n=271\,441$ (!), et en général il y a peu de nombres non-premiers de ce type.
(Ils s'appellent *nombres pseudo-premiers de Perrin*.)
Cette propriété est à la base d'un **test de primalité** probabiliste, similaire au test de Fermat.
