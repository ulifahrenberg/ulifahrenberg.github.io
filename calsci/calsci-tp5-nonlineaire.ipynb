### La mûˋthode du point fixe

La mûˋthode du point fixe consiste û  appliquer l'algorithme itûˋratif suivant :

$$
{\bf x}^{k+1} = {\bf g}({\bf x}^k)
$$

dans le but de rûˋsoudre l'ûˋquation ${\bf g}({\bf x}) = {\bf x}$.
(On l'a dûˋjû  vu pour d'autres cas d'utilisation.)

Suivant le point ${\bf x}^0$ de dûˋpart, la mûˋthode converge ou diverge.

## La mûˋthode de Newton-Raphson

Nous allons appliquer la mûˋthode de Newton pour une fonction vectorielle.
On va dûˋfinir la mûˋthode itûˋrative par analogie avec la mûˋthode 1D. En 1D on a :

$$
x^{k+1} = x^k - \frac{f(x^k)}{f'(x^k)}  \quad \textrm{ou bien} \\
f'(x^k) \; (x^{k+1} - x^k) = - f(x^k)
$$
                               
Cela donne en $n$ dimensions :

$$
J_{\bf f}({\bf x}^k) \; ({\bf x}^{k+1} - {\bf x}^k) =  - {\bf f}({\bf x}^k)
$$

avec $J_{\bf f}$ la matrice Jacobienne de ${\bf f}$ :

$$
J_{\bf f}\left({\bf x}\right)=
\begin{pmatrix} 
\dfrac{\partial f_1}{\partial x_1} & \cdots & \dfrac{\partial f_1}{\partial x_n} \\
\vdots & \ddots & \vdots \\
\dfrac{\partial f_n}{\partial x_1} & \cdots & \dfrac{\partial f_n}{\partial x_n}
\end{pmatrix}
$$

Dans notre cas on veut toujours rûˋsoudre notre systû´me matriciel non linûˋaire
$A({\bf x})\, {\bf x} = {\bf b}$,
nous allons donc prendre

$$
{\bf f}({\bf x}) = A({\bf x})\, {\bf x} - {\bf b}
$$

Notons bien que ${\bf f}$ est une fonction vectorielle û  savoir qu'elle retourne
un vecteur.
